Step 1 : Import Dependencies

make_blobs is used to generate the dataset required for this experiment,
matplotlib is used for data visualization,
and KMeans is used for fitting data based on the K-Means algorithm

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

Step 2 : Generate the dataset

make_blob() is used to generate the dataset for the clustering task by returning the generated dataset and labels 
NOTE : The K-Means algorithm is often used to process unlabeled datasets in production.
The generated dataset contains labels only to facilitate understanding.

```python
X, y = make_blobs(n_samples = 500, n_features = 2, centers = 4, random_state = 1)
```

n_samples is the number of samples,
n_features is the number of features on each sample (the dimension of data),
centers is the number of categories,
and random_state is the seed for the random number generator.
The seed ensures that the data generated every time is the same.

Display the dimension information of the generated dataset.

```python
print("Dimension of X is {} and the dimension of y is {}".format(X.shape, y.shape))
```

The dataset contains 500 sample points, and each sample point contains two features.

Step 3 : Draw the scatter graphs

First draw a scatter graph of the dataset. The labels generated during the dataset generation are not considered for now. The graph contains four clusters


```python
fig, ax1 = plt.subplots(1)
ax1.scatter(X[:, 0], X[:, 1]
            ,marker='o' #Set the shape of the point to circle.
            ,s=8   #Set the size of the point.
           )
plt.show()
```

Next, draw a scatter graph where points in different colors correspond to the generated labels.

```python
color = ["red", "pink", "orange", "green"]

fig, ax1 = plt.subplots(1)

for i in range(4):
    ax1.scatter(X[y==i, 0], X[y==i, 1] #Draw the color based on the label.
            ,marker='o' #Set the shape of the point to circle.
            ,s=8   #Set the size of the point.
            ,c=color[i]
           )
plt.show()
```

Step 4 : Perform k-means clustering

Use sklearn.cluster.KMeans provided by scikit-learn to implement K-Means clustering. First, aggregate the data samples into three types

```python
n_clusters = 3
cluster1 = KMeans(n_clusters = n_clusters, random_state = 3).fit(X)
```

View the label of each sanple point after clustering.

```python
y_pred1 = cluster1.labels_
print(y_pred1)
```

Print the centroid of each cluster

```python
centroid1 = cluster1.cluster_centers_
print(centroid1)
```

Visualize the clustering result

```python
color = ["red", "pink", "orange", "green"]

fig, ax1 = plt.subplots(1)

for i in range(n_clusters):
    ax1.scatter(X[y_pred1==i, 0], X[y_pred1==i, 1] 
            ,marker='o' #Set the shape of the point to circle.
            ,s=8   #Set the size of the point.
            ,c=color[i]
           )

ax1.scatter(centroid1[:, 0], centroid1[:, 1]
            ,marker='x' #Set the shape of the point to cross.
            ,s=15  #Set the size of the point.
            ,c="black"
           )
plt.show()
```

The graph shows that the data samples are aggregated into three clusters, and the centroid of each cluster is represented by x

Step 5 : Perform K-Means Clustering again.

Next, Perform the preceeding steps to use k-means clustering to aggregate data samples into four types.

```python
n_clusters = 4
cluster2 = KMeans(n_clusters = n_clusters, random_state = 0).fit(X)
y_pred2 = cluster2.labels_
centroid2 = cluster2.cluster_centers_
print("Centroid:{}".format(centroid2))
```

Visualize the clustering results

```python
color = ["red", "pink", "orange", "green"]

fig, ax1 = plt.subplots(1)

for i in range(n_clusters):
    ax1.scatter(X[y_pred2==i, 0], X[y_pred2==i, 1] 
            ,marker='o' #Set the shape of the point to circle.
            ,s=8   #Set the size of the point.
            ,c=color[i]
           )

ax1.scatter(centroid2[:, 0], centroid2[:, 1]
            ,marker='x' #Set the shape of the point to cross.
            ,s=15  #Set the size of the point.
            ,c="black"
           )
plt.show()
```
